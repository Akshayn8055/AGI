```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

digits=load_digits()
```

```python
digits.data.shape
```

```python
plt.figure(figsize=(1,1))
plt.imshow(digits.data[1700].reshape(8,8),cmap=plt.cm.gray)
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=0)
X_test.shape

```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

trian
if gamma is too small because it may not capture the complexity of the data if gamma is too large, as it might capture noise in the training data.

```python
svm_clf = SVC(kernel='rbf', gamma='scale')
svm_clf.fit(X_train, y_train)
```

```python
y_pred=svm_clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_estimator(svm_clf,X_test,y_test))
```

```python
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=svm_clf.classes_)
disp.plot(cmap='Blues_r', values_format='d')
#plt.show()
```
